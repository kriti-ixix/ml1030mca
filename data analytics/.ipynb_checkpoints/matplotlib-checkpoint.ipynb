```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'g')
```

```python
plt.plot(y, x, '#ff69b4')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, 'y--')
```

```python
plt.plot(x, y, 'yo')
```

```python
plt.plot(x, y, 'yo')
plt.show()
```

```python
plt.figure(figsize=(10, 12))
plt.plot(x, y, 'g')
plt.savefig('my graph.jpeg')
```
